```python
df_corr = df.corr(numeric_only = True)
#correlation represents how one variable changes as the other one chnnges 
# closer to +1 indicates strong POSITIVE Linear relation
#closer to -1 indicates a strong NEGATIVE linear relation 
# corr = 0 means no relation
df_corr

```

```python
sns.heatmap(df_corr, annot = True)
# annot=True 
#This parameter adds the actual correlation values on top of the heatmap cells. 
#So, each square in the heatmap will display the correlation coefficient 
#between the corresponding pair of variables.
plt.rcParams['figure.figsize'] = (20,7)
plt.show()
```

```python
sns.heatmap(df_corr, annot = True)
plt.rcParams['figure.figsize'] = (25,7)
plt.show()
# to make it readable
```

```python
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only = True).sort_values(by = "2022 Population", ascending = False)
df2
```

```python
df2.plot()
#understanding dataset
```

```python
df3 = df2.transpose()
df3
#making rows columns and vice versa
```
